```python
import torch
torch.cuda.is_available()
```

```python
model_path = "./weights/FastSAM-x.pt"
img_path = "./images/dogs.jpg"
```

```python
from ultralytics import YOLO
from fastsam import FastSAM
from PIL import Image

model = FastSAM(model_path)

input = Image.open(img_path)
input = input.resize((640,640))
input = input.convert("RGB")
everything_results = model(
    input,
    device="cuda",
    retina_masks=True,
    imgsz=640,
    conf=0.4,
    iou=0.9   
    )
```

```python
model.export(format="engine", imgsz=640)
```

```python
from ultralytics.utils import ASSETS, DEFAULT_CFG_DICT, LOGGER, RANK, SETTINGS, callbacks, checks, emojis, yaml_load
from ultralytics.engine.exporter import Exporter

callback = callbacks.get_default_callbacks()
custom = {"imgsz": 640, "batch": 1, "data": None, "verbose": False}  # method defaults
args = {**custom, "mode": "export"}  # highest priority args on the right
Exporter(overrides=args, _callbacks=callback)(model=model)
```
